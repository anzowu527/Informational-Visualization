```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cali = pd.read_csv("california_counties.tsv",sep='\t')
covid = pd.read_csv("covid_surge_offsets.tsv",sep='\t')
```

```python
merge_df = pd.merge(covid, cali, left_on = 'area', right_on='name')
merge_df
```

```python
merge_df['density'] = merge_df['population_2020']/merge_df['area_sq_miles']
merge_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
ax = sns.scatterplot(data=merge_df, x='surge_4_offset', y='density', hue='general_law_or_charter', style='general_law_or_charter', palette='bright')

plt.xlabel('Surge 4 Offset')
plt.ylabel('Population Density')
plt.title('Population Density vs. Surge 4 Offset')
plt.legend(title='General Law or Charter')

# Set y-axis to log scale
ax.set_yscale('log')

plt.show()
```
